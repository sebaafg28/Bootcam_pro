```python
palabra_usuario = input('ingrese su palabra')
for letra_o in palabra_usuario:
    if letra_o == 'o' or letra_o == 'O' :
        print(f"este caracter contiene la letra 'o' :   {letra_o}")
    else: 
        print(f'este caracter no contiene la letra o :  {letra_o}')
```
