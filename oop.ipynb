```python

class Animal :
    especie = 'aguila'
    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color
    def vuelo(self, volar):
        self.volar = volar
        print(f'el {self.especie} mueve su ala {volar} veces y vuela')
    def aterrrizaje(self, aterrizar):
        self.aterrizar = aterrizar
        print(f'abre sus {aterrizar} y aterriza')


aguila = Animal('Aguila', 'Negro')

aguila.vuelo(6)

aguila.aterrrizaje('patas')


```

```python

class Madre :
    
    altura = 160
    piel = 'blanco'
    def __init__(self, belleza):
        self.belleza = belleza
        print(f'su belleza es de {belleza} ')
class Hija(Madre):
    def pelo(self, color):
        self.color = color
        print(f'el color del pelo es {color}')
hijaa = Hija(10)
print(f' la altura es de {hijaa.altura} cm')
print(f'su color de piel es {hijaa.piel}')
hijaa.pelo('rubio')

```
