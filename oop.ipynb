```python

class Animal :
    especie = 'aguila'
    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color
    def vuelo(self, volar):
        self.volar = volar
        print(f'el {self.especie} mueve su ala {volar} veces y vuela')
    def aterrrizaje(self, aterrizar):
        self.aterrizar = aterrizar
        print(f'abre sus {aterrizar} y aterriza')


aguila = Animal('Aguila', 'Negro')

aguila.vuelo(6)

aguila.aterrrizaje('patas')


```
