```python
def obtener_fila (palabra_encontrar, palabra_ingresada):
    cant_de_letras_encontrar = 5

    letras_verificadas = []
    for posicion in range (cant_de_letras_encontrar):
    letras_iguales = palabra_encontrar [posicion] == palabra_ingresada[posicion]
    letra_existente = palabra_ingresada[posicion] in palabra_encontrar[posicion] 
    if letras_iguales :
        letras_verificadas.append ('[' + palabra_ingresada [posicion] + ']' )
    elif letra_existente:
        letras_verificadas.append('('+ palabra_ingresada[posicion]+ ']')
    else: 
        letras_verificadas.append(palabra_ingresada[posicion])
    return letras_verificadas
resultado = obtener_fila(input('Ingrese su palabra secreta', input('ingrese su palabra para encontar'))) 
```
